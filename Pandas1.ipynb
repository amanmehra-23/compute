```python

df_filt = df[df['duration_ms'] < 400000] # this a filter as we proceed to fint the most energetic songs
df_filt
```

```python
df_filt2 = df_filt.sort_values('energy')[['song_title', 'artist', 'duration_ms','energy']].tail(10)
#filtering songs based on enegry out of the ones we filter earlier and finding the most energetic ones
df_filt2
```

```python
ans = df_filt2.sort_values('energy', ascending=False)
#changing the order to descending now
ans
```
